```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =  pd.read_csv("/kaggle/input/fastfood-nutrition/fastfood.csv")
```

```python
df.head()
```

```python
x = df[["total_fat", "total_carb", "sodium"]]
y= df["calories"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
import numpy as np
print(np.isnan(x_train).any())
print(np.isinf(x_train).any())
print(np.isnan(y_train).any())
print(np.isinf(y_train).any())
```

```python
x_train = np.nan_to_num(x_train)
y_train = np.nan_to_num(y_train)
```

```python
from sklearn.linear_model import LinearRegression
mlr= LinearRegression()
mlr.fit(x_train,y_train)
```

```python
print("intercept:", mlr.intercept_)
print("coefficiants:")
list(zip(x,mlr.coef_))
```

```python
y_pred = mlr.predict(x_test)
print("prediction for test set:{}".format(y_pred))
```

```python
mlr_dif = pd.DataFrame({"actual value: ":y_test,"predited value: ":y_pred})
mlr_dif.head()
```

```python
from sklearn import metrics
meansquare = metrics.mean_squared_error(y_test, y_pred)
print("R-squared: {:.2f}".format(mlr.score(x, y) * 100))
print("Mean Squared Error: ", meansquare)
```
