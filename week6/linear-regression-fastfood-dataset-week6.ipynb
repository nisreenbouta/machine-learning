# **Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df = pd.read_csv("/kaggle/input/fastfood-nutrition/fastfood.csv")
```

```python
df 
```

```python
df.head()  
```

```python
plt.scatter(df.cal_fat,df.calories) 
plt.xlabel("cal_fat") 
plt.ylabel("calories")    
plt.show()   
```

```python

x = df.cal_fat.values      
```

```python
x.shape 
```

```python
x = df.cal_fat.values.reshape(-1,1)
y = df.calories.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression() 
linear_reg.fit(x,y)
```

```python
#prediction
import numpy as np

b0 = linear_reg.predict([[0]]) 
print("b0: ",b0)              
```

```python
b0_ = linear_reg.intercept_   
print("b0: ",b0_)   
```

```python
b1 = linear_reg.coef_  
print("b1: ",b1)   
```

```python
b11 = linear_reg.predict([[11]]) 
print("b11: ",b11)
```

```python
y_head = linear_reg.predict(x)
```

```python
plt.plot(x, y_head, color="pink")
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.metrics import r2_score
print("R Square Score: ",r2_score(y,y_head))
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared Error: ",mean_squared_error(y,y_head))
```

```python
MSE = np.square(np.subtract(y,y_head)).mean()
print("Mean Squared Error: ",MSE)
```
